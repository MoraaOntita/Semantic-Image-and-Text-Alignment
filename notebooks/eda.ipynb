### Text Analysis

```python
# List unique concepts
unique_concepts = concepts_df['concept'].unique()

# Print the unique concepts
print("Unique Concepts:")
for concept in unique_concepts:
    print(concept)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the descriptions to TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(concepts_df['concept'])

# Compute cosine similarity matrix
cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Display similarity matrix (or you can print or analyze specific parts of it)
print("Cosine Similarity Matrix:")
print(cosine_sim_matrix)

# Optional: Identify pairs with high similarity (e.g., similarity > 0.8)
import numpy as np

# Find pairs with high similarity
high_sim_indices = np.where(cosine_sim_matrix > 0.8)
for i, j in zip(*high_sim_indices):
    if i < j:  # Avoid printing the same pair twice
        print(f"Concepts {i} and {j} are similar with a score of {cosine_sim_matrix[i, j]:.2f}")
```

- 1 indicates perfect similarity (the two concepts are identical in terms of their TF-IDF vector representation).
- 0 indicates no similarity.

**Example Analysis:**

*Concepts 3 and 37:* These concepts are identical (similarity score of 1.00), meaning they are exactly the same in terms of their content or description.

*Concepts 34 and 50:* These concepts have a similarity score of 0.81, indicating a high degree of similarity but not exact duplication. This suggests these concepts share a lot of content or wording but have some differences.

**High Similarity Pairs**
The list of similar concepts highlights pairs with high similarity scores, indicating redundancy or overlap in the descriptions. Here’s what you can infer:

*Exact Matches (Score of 1.00):* Concepts listed with a similarity score of 1.00 are exact duplicates or very close in content. This may need to be investigated to determine if they should be merged or if they indeed represent the same concept.

*High Similarity (Score ~0.81 - 0.83):* These pairs have high similarity but are not identical. It may be useful to review these concepts to understand their differences and see if the similarity is due to overlapping content or themes.

```python
# Example DataFrame for asset suggestions
# Replace 'asset_suggestions' with your actual column name
asset_suggestions = concepts_df['asset_suggestions'].dropna()

# Combine all asset suggestions into a single list
all_assets = ' '.join(asset_suggestions)

# Tokenize and count frequency of each asset suggestion
from collections import Counter
import re

# Tokenize the asset suggestions (basic tokenization by spaces)
tokens = re.findall(r'\b\w+\b', all_assets.lower())  # Convert to lower case for uniformity

# Count frequency of each asset suggestion
asset_counts = Counter(tokens)

# Print most common asset suggestions
print("Most Common Asset Suggestions:")
for asset, count in asset_counts.most_common():
    print(f"{asset}: {count}")
```
